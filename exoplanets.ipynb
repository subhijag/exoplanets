```python
sns.set(style="whitegrid")
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/planets.csv"
df = pd.read_csv(url)
```

```python
print("First 5 rows:")
display(df.head())
```

```python
print("\nDataset info:")
df.info()
```

```python
df = df.dropna()
```

```python
plt.figure()
sns.scatterplot(data=df, x="orbital_period", y="mass")
plt.xscale("log")
plt.yscale("log")
plt.title("Exoplanet Mass vs Orbital Period")
plt.xlabel("Orbital Period (days)")
plt.ylabel("Mass (Jupiter mass)")
plt.show()
```

```python
plt.figure()
df["method"].value_counts().plot(kind="bar")
plt.title("Exoplanet Discovery Methods")
plt.xlabel("Method")
plt.ylabel("Number of Planets")
plt.show()
```

```python
print("\nAverage planet mass by discovery method:")
display(df.groupby("method")["mass"].mean().sort_values(ascending=False))
```
